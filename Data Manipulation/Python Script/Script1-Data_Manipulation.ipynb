```python
df = pd.read_excel(pwd + "\\Data - Survey Monkey Output.xlsx ", sheet_name = "Sheet1")
df
```

```python
#Making a copy of my dataset:

df_modified = df.copy()
df_modified.columns
```

```python
columns_to_drop = ['Start Date-', 'End Date-', 'Email Address-',
                   'First Name-', 'Last Name-', 'Custom Data 1-']
df_modified = df_modified.drop(columns= columns_to_drop)
df_modified
```

```python
#Here, we are gonna use the melt() function:

id_vars = list(df_modified.columns)[:8]
val_vars = list(df_modified.columns)[8:]
```

```python
df_melted = df_modified.melt(id_vars=id_vars, value_vars=val_vars, var_name='Question-Subquestion', value_name='Answer')
df_melted
```

```python
df_questions = pd.read_excel(pwd+ "\\Data - Survey Monkey Output.xlsx", sheet_name = "Edited_question")
df_questions
```

```python
df_modify_questions = df_questions.copy()
df_modify_questions
```

```python
df_modify_questions.columns
```

```python
df_modify_questions = df_modify_questions.drop(columns = ['Raw_Question', 'Raw_Subquestion','Subquestion'])
df_modify_questions
```

```python
df_merged = pd.merge(left=df_melted, right=df_modify_questions, how="left", left_on="Question-Subquestion", right_on="Question + Subquestion")
print("Original Data", len(df_melted))
print("Merged Data", len(df_merged))
df_merged
```

```python
df_merged = df_merged.drop(columns= 'Question + Subquestion')
df_merged
```

```python
count_respondents = df_merged[df_merged["Answer"].notna()]
count_respondents = count_respondents.groupby("Question")["Respondent ID-"].nunique().reset_index()
count_respondents.rename(columns={"Respondent ID-":"Respondents"}, inplace = True)
count_respondents
```

```python
df_merged1 = pd.merge(left=df_merged, right=count_respondents, how="left", left_on="Question", right_on="Question")
print("Original Data", len(df_merged))
print("Merged Data", len(df_merged1))
df_merged1
```

```python
same_answer_count = df_merged #[df_merged["Answer"].notna()]
same_answer_count = same_answer_count.groupby(["Question-Subquestion", "Answer"])["Respondent ID-"].nunique().reset_index()
same_answer_count.rename(columns={"Respondent ID-":"Same Answer"}, inplace = True)
same_answer_count
```

```python
df_merged2 = pd.merge(left=df_merged1, right=same_answer_count, how="left", left_on=["Question-Subquestion","Answer"], right_on=["Question-Subquestion","Answer"])
df_merged2["Same Answer"].fillna(0, inplace = True)
print("Original Data", len(df_merged1))
print("Merged Data", len(df_merged2))
df_merged2
```

```python
#Renaming the columns of the final output data:

final_output_data = df_merged2.copy()
final_output_data.rename(columns= {"Identify which division you work in.-Response": "Primary Division",	"Identify which division you work in.-Other (please specify)" : "Secondary Division","Which of the following best describes your position level?-Response" : "Level of Position",
                                   "Which generation are you apart of?-Response" : "Generation", "Please select the gender in which you identify.-Response": "Gender", "Which duration range best aligns with your tenure at your company?-Response": "Tenure",
                                   "Which of the following best describes your employment type?-Response": "Employment Type"}, inplace= True)
final_output_data

```
